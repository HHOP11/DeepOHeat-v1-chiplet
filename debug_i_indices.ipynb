# Debug i1..i6 for 7-layer z interfaces

This notebook verifies interface indices mapped from `stack.z_interfaces_norm`.

```python
import numpy as np
import heat_chiplet as hc
```

```python
# Same nz rule as training data generation in heat_chiplet.py
nz = max(2, int(round(hc.stack.total_thickness / hc.stack.min_sampling_spacing_m)) + 1)
z_if_star = hc.stack.z_interfaces_norm[1:-1]
idx = [int(round(v * (nz - 1))) for v in z_if_star]

z_grid = np.linspace(0.0, 1.0, nz)

print('total_thickness:', hc.stack.total_thickness)
print('min_sampling_spacing_m:', hc.stack.min_sampling_spacing_m)
print('nz:', nz)
print('z_if_star:', z_if_star)
print('i1..i6:', idx)
print('z_grid[idx]:', z_grid[idx])
print('abs diff:', np.abs(z_grid[idx] - z_if_star))
```

```python
# Optional: verify with a custom nz
nz_custom = 56
idx_custom = [int(round(v * (nz_custom - 1))) for v in z_if_star]
z_grid_custom = np.linspace(0.0, 1.0, nz_custom)

print('nz_custom:', nz_custom)
print('i1..i6 (custom):', idx_custom)
print('z_grid_custom[idx_custom]:', z_grid_custom[idx_custom])
print('abs diff (custom):', np.abs(z_grid_custom[idx_custom] - z_if_star))
```
